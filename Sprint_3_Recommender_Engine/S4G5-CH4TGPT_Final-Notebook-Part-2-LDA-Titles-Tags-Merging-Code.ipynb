```python
import pandas as pd
```

```python
tags = pd.read_csv("lda_tag_topics.csv")
titles = pd.read_csv("lda_title_topics.csv")
```

```python
topic_model_df = titles.merge(tags, how="left", on="Unnamed: 0")
topic_model_df
```

```python
topic_model_df.drop(columns=["video_title", "video_tags"], inplace=True)
topic_model_df
```

```python
main_df = pd.read_csv("main_df.csv", sep="|")
```

```python
len(main_df)
```

```python
main_df = main_df.merge(topic_model_df, how="left", on="Unnamed: 0")
```

```python
main_df.to_csv("conso_df.csv", sep="|")
```
